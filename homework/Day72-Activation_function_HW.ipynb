# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```

```python
def relu(x):
  return max(x, 0)
```

```python
def drelu(x):
  if x > 0:
    return 1
  else:
    return 0 
```

### Draw ReLU and DReLU

```python
x = plt.linspace(-10, 10, 100)
y = np.array([relu(t) for t in x])
z = np.array([drelu(t) for t in x])
plt.plot(x, y, 'b', label = 'relu')
plt.plot(x, z, 'r', label = 'drelu')
plt.grid()
plt.text(-2.5, 8, '$relu(x) = max(x, 0)$', fontsize = 15)
plt.text(-10, 1, r'$drelu(x) = \frac {d}{dx} relu(x)$', fontsize = 15)
plt.title('ReLU and DReLU function')

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))

plt.show()
```
