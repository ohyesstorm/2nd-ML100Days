```python
def gradient_decent(df, lr, cur_x, iters = 0, prev_step_size = 1, precision = 0.000001, max_iters = 10000):
  iters_history = [iters]
  precision_history = [prev_step_size]
  while prev_step_size > precision and iters < max_iters:
    prev_x = cur_x
    cur_x = cur_x - lr*df(cur_x)
    prev_step_size = abs(cur_x-prev_x)
    iters = iters + 1
    print('iters:', iters)
    iters_history.append(iters)
    precision_history.append(prev_step_size)
  return iters_history, precision_history
```

```python
df = lambda x: 2*(x+5) #Gradient of our function 
iter_his_1, precision_his_1 = gradient_decent(df, lr=0.1, cur_x=3)
```

```python
iter_his_2, precision_his_2 = gradient_decent(df, lr = 0.0001, cur_x=3)
```

```python
plt.plot(iter_his_1, precision_his_1, 'b')
# plt.plot(iter_his_2, precision_his_2, 'r')
plt.show()
```
